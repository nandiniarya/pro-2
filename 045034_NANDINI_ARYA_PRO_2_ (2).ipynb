```python
Bank_inputs_names = Bank_inputs.columns; Bank_inputs_names
Bank_outputs_labels = Bank_outputs['cluster_number_x'].unique().astype(str); Bank_outputs_labels
```

```python
# Initialize StratifiedShuffleSplit with desired test size and random state
stratified_split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=45034)
# Perform the stratified split to get training and testing indices
for i, j in stratified_split.split(Bank_inputs, Bank_outputs):
    train_Bank_inputs, test_Bank_inputs = Bank_inputs.iloc[i], Bank_inputs.iloc[j]
    train_Bank_outputs, test_Bank_outputs = Bank_outputs.iloc[i], Bank_outputs.iloc[j]
```

```python
def memory_usage():
    """
    Function to get memory usage.
    """
    process = psutil.Process()
    mem = process.memory_info().rss
    return mem / (1024 * 1024)  # Convert to MB
```
